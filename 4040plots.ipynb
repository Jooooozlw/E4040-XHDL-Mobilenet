```python
table_7 = ['cifar100_mn_alpha_1_size224_20191202-172921.log', 'cifar100_mn_alpha_1_size192_20191202-163053.log', 
           'cifar100_mn_alpha_1_size160_20191202-154908.log', 'cifar100_mn_alpha_1_size128_20191202-152329.log']
title_t7 = ['Table 7 -- Accuracy', 'Table 7 -- Loss', 
            'Table 7 -- Validation Accuracy', 'Table 7 -- Validation Loss']
y_label_t7 = ['accuracy', 'loss', 'accuracy', 'loss']
y_t7 = ['accuracy' ,'loss', 'val_accuracy', 'val_loss']
legend_t7 = ['224', '192', '160', '128']

fig, ax = plt.subplots(2, 2)

i = 0
for row in ax:
    for col in row:
        
        col.plot(files[table_7[0]].epoch, files[table_7[0]][y_t7[i]])
        col.plot(files[table_7[1]].epoch, files[table_7[1]][y_t7[i]])
        col.plot(files[table_7[2]].epoch, files[table_7[2]][y_t7[i]])
        col.plot(files[table_7[3]].epoch, files[table_7[3]][y_t7[i]])
        
        col.legend(legend_t7)
        col.set_title(title_t7[i])
        col.set_xlabel("epoches")
        col.set_ylabel(y_label_t7[i])
        i += 1
       
plt.subplots_adjust(hspace = 0.8)
plt.show()
```
